```python
import torch as pt
```

```python
# Empty matrix
x=pt.empty(3,3)
x
```

```python
# Randomly initialized matrix
x=pt.rand(3,3)
x
```

```python
# Zeros matrix
x=pt.zeros(3,3,dtype=pt.long)
x
```

```python
# Construct a tensor directly from data
x=pt.tensor([3.5,3])
x
```

```python
# size
print(x.size())
```

```python
# Changing numpy array to torch
import numpy as np
a=np.ones(5)
b=pt.from_numpy(a)
print(a)
print(b)
```
