```python


import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import texthero as hero # text hero  
```

```python
yt = pd.read_csv("../input/youtube-videos-dataset/youtube.csv")
```

```python
yt.head(3)
```

```python
yt.shape
```

Create PCA column for the YouTube dataset

```python
yt["pca"] = (
            yt["title"]
            .pipe(hero.clean)
            .pipe(hero.tfidf) # term frequency and inverse document frequency 
            .pipe(hero.pca)
)
```

```python
yt.head()
```

```python
hero.scatterplot(
    yt, 
    col='pca', 
    color='category', 
    title="PCA YouTube Video Title"
)
```
